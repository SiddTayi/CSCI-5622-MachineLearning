```python
import matplotlib.pyplot as plt
import numpy as np

# Given points and their target outputs
points = np.array([[-2, 4], [2, 2], [4, 2]])
targets = np.array([-1, 1, 1])

# Separating line equation coefficients
a, b, c = 0, -1, 3

# Function to calculate y values for the line based on x values
def line_equation(x):
    return (a * x + c) / -b

# Scatter plot for the points
plt.scatter(points[:, 0], points[:, 1], c=targets, cmap='viridis', marker='o')

# Plotting the separating line
x_values = np.linspace(-3, 5, 100)
y_values = line_equation(x_values)
plt.plot(x_values, y_values, '--', color='red', label='Separating Line')

# Highlighting the support vectors
support_vectors = points[targets == 1]
plt.scatter(support_vectors[:, 0], support_vectors[:, 1], s=150, facecolors='none', edgecolors='red', linewidths=2, label='Support Vectors')

# Setting labels and legend
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Support Vector Machine')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Show plot
plt.show()
```
