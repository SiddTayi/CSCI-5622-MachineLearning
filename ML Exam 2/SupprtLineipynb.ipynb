```python
import numpy as np
import matplotlib.pyplot as plt

# Given weight vector and bias term
w = np.array([-5/18, -7/36])
b = 35/18

# Given points and classes
X1 = np.array([-2, 4])
y1 = -1

X2 = np.array([2, 2])
y2 = 1

X3 = np.array([4, 2])
y3 = 1

# Scatter plot for points
plt.scatter(X1[0], X1[1], c='r', marker='o', label='Class -1')
plt.scatter(X2[0], X2[1], c='b', marker='o', label='Class 1')
plt.scatter(X3[0], X3[1], c='b', marker='o')

# Plotting the decision boundary line
x_values = np.linspace(-3, 5, 100)
y_values = (-10/7) * x_values + 5
plt.plot(x_values, y_values, label='Decision Boundary', linestyle='--')

# Adding labels and legend
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('SVM Decision Boundary')
plt.legend()

# Show the plot
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```
